```python
numerical_features = df.select_dtypes(include=['number'])
```

```python
corr = numerical_features.corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='icefire')
plt.show()
```

```python
fig = plt.figure(figsize=(8, 5))  #Now plt is defined and can be used.
sns.barplot(x="UN_Region", y="IMF_Estimate", data=df, errorbar=None)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['WorldBank_Estimate'], df['UN_Estimate'], color='blue', alpha=0.6)

plt.title('UN Estimates vs World Bank Estimates')
plt.xlabel('World Bank Estimate')
plt.ylabel('UN Estimate')

plt.grid(True)

import numpy as np

m, b = np.polyfit(df['WorldBank_Estimate'], df['UN_Estimate'], 1)
plt.plot(df['WorldBank_Estimate'], m * df['WorldBank_Estimate'] + b, color='orange', linestyle='-')

plt.scatter(df['WorldBank_Estimate'], df['UN_Estimate'], color='green')

plt.show()
```
