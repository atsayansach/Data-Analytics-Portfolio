```python
plt.figure(figsize=(10, 6))
plt.scatter(df['WorldBank_Estimate'], df['UN_Estimate'], color='blue', alpha=0.6)

plt.title('UN Estimates vs World Bank Estimates')
plt.xlabel('World Bank Estimate')
plt.ylabel('UN Estimate')

plt.grid(True)

import numpy as np

m, b = np.polyfit(df['WorldBank_Estimate'], df['UN_Estimate'], 1)
plt.plot(df['WorldBank_Estimate'], m * df['WorldBank_Estimate'] + b, color='orange', linestyle='-')

plt.scatter(df['WorldBank_Estimate'], df['UN_Estimate'], color='green')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
counts, bins, patches = plt.hist(df['mark'], bins=10, edgecolor='black')

colors = ['#FF5733', '#FFC300', '#DAF7A6', '#33FFBD',
          '#33A1FF', '#9B59B6', '#E91E63', '#2ECC71',
          '#F39C12', '#1ABC9C']  # 10 colours for 10 bins

for patch, color in zip(patches, colors):
    patch.set_facecolor(color)

plt.xlabel('Mark')
plt.ylabel('Frequency')
plt.title('Marks histogram with different colours per bin')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
counts, bins, patches = plt.hist(df['mark'], bins=10, edgecolor='black')

# 10 simple Matplotlib colour names
colors = ['red', 'orange', 'gold', 'yellowgreen', 'green',
          'turquoise', 'blue', 'slateblue', 'purple', 'pink']

for patch, color in zip(patches, colors):
    patch.set_facecolor(color)

plt.xlabel('Mark')
plt.ylabel('Frequency')
plt.title('Marks histogram with different colours per bin')
plt.show()
```

```python
# Count by Gender
ax = sns.countplot(x='gender', data=df)
# Change code to : ax = sns.countplot(x='gender', data=df, palette='husl') to change colour palette.
ax.bar_label(ax.containers[0])
plt.title("# of Students by Gender")
plt.show()
```

```python
ax = sns.countplot(
    x="gender",
    data=df[df['gender'].notna()],  # Exclude missing gender values
    palette="Set2"  # Different colours per bar
)
ax.bar_label(ax.containers[0])
plt.title("# of Students by Gender")
plt.show()
```

```python
# Count by Gender with matplotlib named colors
ax = sns.countplot(x="gender", data=df[df['gender'].notna()])

# Change each bar color using named colors
colors = ['red', 'blue']  # female, male
for patch, color in zip(ax.patches, colors):
    patch.set_color(color)

ax.bar_label(ax.containers[0])
plt.title("# of Students by Gender")
plt.show()
```
